```python
import networkx as nx

def readFile(path: str) -> list:
    coords: list = []
    edges: list = []
    with open(path, "r", encoding="utf-8") as file:
        for lineIdx, line in enumerate(file.readlines()):
            splittedData = [int(val) for val in line.split(" ")[:-1]]
            coords.append((splittedData[1], splittedData[2]))
            for _, nodeIdx in enumerate(splittedData[3::]):
                edges.append((lineIdx, nodeIdx))
    return coords, edges

def readPathFromFile(path: str) -> list:
    nodes: list = []
    with open(path, "r", encoding="utf-8") as file:
        for lineIdx, line in enumerate(file.readlines()):
            # print(line.split(" "))
            nodes = [int(node) for node in line.split(" ")[:-1]]
    return nodes

def drawGraph(G: nx.Graph, coords: list, edges: list, node_size=20) -> None:
    for node in range(len(coords)):
        G.add_node(node, pos=coords[node])
    G.add_edges_from(edges)
    pos = nx.get_node_attributes(G, 'pos')
    nx.draw(G, pos, node_size=node_size)
    G.remove_nodes_from(list(range(len(coords))))
```
